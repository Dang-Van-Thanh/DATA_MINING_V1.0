# Association Rule Mining

```python
from sklearn.preprocessing import OneHotEncoder
from src.mining.association import AssociationMiner

assoc_df = df_clean[cat_cols + ["y"]].dropna()

encoder = OneHotEncoder(sparse=False)
assoc_matrix = encoder.fit_transform(assoc_df)
assoc_encoded = pd.DataFrame(
    assoc_matrix,
    columns=encoder.get_feature_names_out()
)

miner = AssociationMiner(
    min_support=0.05,
    min_confidence=0.6,
    min_lift=1.2
)

rules = miner.run(assoc_encoded)
rules.head(10)
```

Insight mẫu: housing=yes & loan=no → y=yes (lift cao).

# Clustering

```python
from src.mining.clustering import ClusteringModel

clusterer = ClusteringModel(k_range=[3,4,5,6])
best_k, scores = clusterer.find_best_k(X_transformed)

best_k, scores
```

```python
model, labels = clusterer.fit(X_transformed, best_k)
df_clean["cluster"] = labels

df_clean.groupby("cluster")[num_cols].mean()
```

Insight: mỗi cụm đại diện 1 hồ sơ tài chính khác nhau.

```python
from sklearn.metrics import silhouette_score

sil_scores = {}
for k in [3,4,5,6]:
    model, labels = clusterer.fit(X_transformed, k)
    sil_scores[k] = silhouette_score(X_transformed, labels)

pd.DataFrame.from_dict(
    sil_scores, orient="index", columns=["silhouette"]
)
```

```python
plt.plot(list(sil_scores.keys()), list(sil_scores.values()), marker="o")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette score")
plt.title("Clustering evaluation")
plt.show()
```

- k = 4 cho silhouette cao nhất → cụm tách biệt tốt
- Các cụm phản ánh hồ sơ tài chính khác nhau
- Có thể dùng cluster làm feature cho marketing
